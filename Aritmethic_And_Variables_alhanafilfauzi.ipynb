# Variabel (Variables)
Sejauh ini, Anda telah menggunakan kode untuk membuat kalkulasi dan mencetak hasilnya, dan hasilnya tidak disimpan di mana pun. Namun, Anda dapat membayangkan bahwa Anda mungkin ingin menyimpan hasilnya untuk digunakan nanti. Untuk itu, Anda perlu menggunakan **variabel**.

# Membuat Variabel
**Cell code** berikutnya membuat variabel bernama **test_var** dan memberikannya nilai yang kita dapatkan ketika kita menambahkan 5 ke 4.
Kita kemudian mencetak nilai yang ditetapkan ke variabel tersebut, yaitu 9.

```python
# Membuat sebuah variabel dengan nama test_var dan berikan nilai 4+5
test_var = 4 + 5

# Cetak nilai dari variabel test_var
print(test_var)
```

Secara umum, sebelum menentukan nama variabel, Anda harus mulai dengan memilih nama yang ingin Anda gunakan. Nama variabel idealnya pendek dan deskriptif. Mereka juga harus memenuhi beberapa persyaratan:

1. Tidak boleh ada spasi (misalnya, **test var** tidak diperbolehkan)
2. Mereka hanya boleh menyertakan huruf, angka, dan garis bawah (misalnya, **test_var!** tidak diperbolehkan)
3. Nama variabel harus dimulai dengan huruf atau garis bawah (misalnya, **1_var** tidak diperbolehkan)

Kemudian, untuk membuat variabel, Anda perlu menggunakan **=** untuk menetapkan nilai yang Anda inginkan.
Anda selalu dapat melihat nilai yang ditetapkan ke variabel dengan menggunakan **print()** dan meletakkan nama variabel dalam tanda kurung.
Seiring berjalannya waktu, Anda akan belajar bagaimana memilih nama yang baik untuk variabel Python. Tidak masalah jika sekarang terasa tidak nyaman, dan cara terbaik untuk belajar adalah dengan melihat banyak kode Python!

# Mengolah Variabel
Anda selalu dapat mengubah nilai yang ditetapkan ke variabel dengan menimpa nilai sebelumnya Pada sel kode di bawah ini, kami mengubah nilai **my_var** dari 3 menjadi 100.

```python
# Menetapkan nilai variabel my_var dengan 3
my_var = 3

# Mencetak nilai yang sudah ditetapkan
print(my_var)

# Mengganti nilai variabel my_var menjadi 100
my_var = 100

# Mencetak Nilai yang sudah diubah
print(my_var)
```

Perhatikan bahwa secara umum, setiap kali Anda mendefinisikan variabel dalam sel kode, semua sel kode berikutnya juga memiliki akses ke variabel tersebut. Sebagai contoh, kita menggunakan sel kode berikutnya untuk mengakses nilai **my_var** (dari **cell code** di atas) dan **test_var**

```python
print(my_var)
print(test_var)
```

**Cell code** berikutnya memberitahu Python untuk meningkatkan nilai **my_var** saat ini sebesar 3.
Untuk melakukan ini, kita masih perlu menggunakan **my_var =** seperti sebelumnya. Dan juga seperti sebelumnya, nilai baru yang ingin kita tetapkan ke variabel berada di sebelah kanan tanda** =.**

```python
# Tambahkan nilai dengan 3
my_var = my_var + 3

# Cetak nilai setelah ditambahkan
print(my_var)
```

# Menggunakan Beberapa Variabel
Adalah hal yang umum bagi kode untuk menggunakan banyak variabel. Hal ini sangat berguna ketika kita harus melakukan perhitungan yang panjang dengan beberapa input.
Pada **cell code** berikutnya, kita menghitung jumlah detik dalam empat tahun. Perhitungan ini menggunakan lima input.

```python
# Membuat Variabel
num_years = 4
days_per_year = 365 
hours_per_day = 24
mins_per_hour = 60
secs_per_min = 60

# Hitung jumlah detik
total_secs = secs_per_min * mins_per_hour * hours_per_day * days_per_year * num_years
print(total_secs)
```

Seperti yang dihitung di atas, ada **126144000 detik** dalam empat tahun.
Perlu dicatat bahwa perhitungan ini dapat dilakukan tanpa variabel hanya dengan **60 * 60 * 24 * 365 * 4**, namun akan jauh lebih sulit untuk memeriksa bahwa perhitungan tanpa variabel tidak memiliki kesalahan, karena tidak mudah dibaca. Ketika kita menggunakan variabel (seperti **num_years, days_per_year**, dll), kita dapat melacak setiap bagian dari kalkulasi dengan lebih baik dan lebih mudah untuk memeriksa dan memperbaiki kesalahan.
Perhatikan bahwa sangat berguna untuk menggunakan variabel ketika nilai dari input dapat berubah. Sebagai contoh, katakanlah kita ingin sedikit meningkatkan perkiraan kita dengan memperbarui nilai jumlah hari dalam setahun dari 365 menjadi 365,25, untuk memperhitungkan tahun kabisat. Kemudian kita bisa mengubah nilai yang diberikan ke **days_per_year** tanpa mengubah variabel lainnya dan mengulang perhitungan.

```python
# Perbaharui
days_per_year = 365.25

# Hitung jumlah detik dalam 4 tahun
total_secs = secs_per_min * mins_per_hour * hours_per_day * days_per_year * num_years
print(total_secs)
```

**Catatan:** Anda mungkin telah memperhatikan angka .0 yang ditambahkan di akhir angka, yang mungkin terlihat tidak perlu. Hal ini disebabkan oleh fakta bahwa pada perhitungan kedua, kami menggunakan angka dengan bagian pecahan (365,25), sedangkan perhitungan pertama hanya mengalikan angka tanpa bagian pecahan.
